```python
import pandas as pd
# Create a DataFrame
data = {
    'Age': [23, 45, 12, 35, 46],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Salary': [70000, 80000, 15000, 50000, 90000]
}
df = pd.DataFrame(data)
print("DataFrame:")
print(df)
# Data manipulation
# 1. Selecting a column
print("\nAges:")
print(df['Age'])
# 2. Filtering rows
print("\nFiltered DataFrame (Age > 30):")
filtered_df = df[df['Age'] > 30]
print(filtered_df)
# 3. Adding a new column
df['Tax'] = df['Salary'] * 0.1
print("\nDataFrame with Tax column:")
print(df)
# 4. Grouping and aggregation
print("\nMean Salary by Age Group:")
print(df.groupby('Age')['Salary'].mean())
# 5. Handling missing values
df_with_nan = df.copy()
df_with_nan.loc[2, 'Salary'] = None  # Introduce a NaN value
print("\nDataFrame with NaN:")
print(df_with_nan)
print("\nDataFrame after filling NaN with median salary:")
df_with_nan['Salary'] = df_with_nan['Salary'].fillna(df_with_nan['Salary'].median())
print(df_with_nan)
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Create sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
# Line plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Line Plot')
plt.title('Line Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
# Bar plot
plt.figure(figsize=(10, 6))
plt.bar(x, y, color='green', alpha=0.7, label='Bar Plot')
plt.title('Bar Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
# Histogram
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
plt.figure(figsize=(10, 6))
plt.hist(data, bins=4, color='purple', alpha=0.7, rwidth=0.85)
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='red', label='Scatter Plot')
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
# Box plot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.figure(figsize=(10, 6))
plt.boxplot(data, vert=True, patch_artist=True)
plt.title('Box Plot Example')
plt.xlabel('Distribution')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```
